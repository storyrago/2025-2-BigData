```python
import matplotlib.pyplot as plt
import pandas as pd

df = pd.read_csv("학생건강검사.csv", encoding="cp949")
students = df.to_numpy()
elementCnt = 0
middleCnt = 0
highCnt = 0
elementHeight = 0
middleHeight = 0
highHeight = 0
for i in range(1,len(students)):
  if students[i][1] == "초":
    elementHeight += students[i][4]
    elementCnt+=1
  if students[i][1] == "중":
    middleHeight += students[i][4]
    middleCnt+=1
  if students[i][1] == "고":
    highHeight += students[i][4]
    highCnt+=1
elementAvg = elementHeight / elementCnt
middleAvg = middleHeight / middleCnt
highAvg = highHeight / highCnt
x = ['element','middle','high']
y = [elementAvg,middleAvg,highAvg]
plt.bar(x,y,label="AvgHeightBar")
plt.title("Average Height")
plt.xlabel("Grade")
plt.ylabel("Average Height")
plt.legend(loc="upper left")

plt.show()
```

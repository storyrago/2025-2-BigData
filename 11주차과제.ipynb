```python
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pylab as plt
import pandas as pd
```

```python
df = pd.read_csv("CARD_SUBWAY_MONTH_202508.csv", encoding='utf-8')

a = df.values.tolist()
an = np.array(a);
print(an)
```

```python
bn = np.array([])
x = np.array([])
y1 = np.array([])
cnt = 0
for i in an:
  if i[1] == '한성대입구(삼선교)':
    cnt += 1
    x = np.append(x,cnt)
    bn = np.append(bn,i)
bn = bn.reshape(-1,6)
for i in bn:
  y1 = np.append(y1,int(i[3])) #하차총승객
x = x.reshape(-1,1)

model2 = LinearRegression()
model2.fit(x,y1)

#예측
y1_pred = model2.predict(x)

#산점도
plt.scatter(x, y1, color='green', alpha=0.6, label='data')

#시각화
plt.plot(x, y1_pred, color='orange', linewidth=2, label='LinearRegression')
plt.xlabel('data sequence')
plt.ylabel('alighting')
plt.title('alighting LinearRegression')

#범례 표시
plt.legend()

plt.show()
```
